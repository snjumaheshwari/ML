### Bar Chart

```python
numbers= np.random.randint(0,10,5)
indices= np.arange(5)
print(numbers)
print(indices)
```

```python
plt.bar(indices,numbers)
plt.show()
```

```python
plt.bar(indices,numbers,0.5,color='orange',label='gold')
plt.legend()
plt.show()
```

```python
numbers=np.random.randint(0,10,5)
indices=np.arange(5)

print(numbers)
print(indices)
```

```python
plt.bar(indices,numbers)
plt.show()
```

```python
plt.bar(indices,numbers,0.5,color='green',label="success")
plt.legend()
plt.show()
```

```python
numbers2=np.random.randint(1,10,5)
plt.bar(indices,numbers2,0.25,color='red',label='fail')
plt.bar(indices+0.25,numbers,0.25,color='green',label='success')
plt.legend()
plt.show()
```

### Pie Chart

```python
labels=['AC','TLE','Runtime Error',
        'Compilation Error','Partially accepted','WA']

values=[100,20,39,20,54,120]
plt.pie(values,labels=labels)
plt.show()
```

```python
labels=['AC','TLE','Runtime Error',
        'Compilation Error','Partially accepted','WA']

values=[100,20,39,20,54,120]
plt.pie(values,labels=labels, radius=1.5,explode=(0,1,0,0,0,0), shadow=True)
plt.show()
```

### Subplots

```python
x=np.arange(100)
y=x**2

x1=np.arange(50,200)
y1=x1+20
plt.scatter(x,y)
plt.scatter(x1,y1)
plt.show()
```

```python
plt.figure()
plt.scatter(x,y)
plt.figure()
plt.scatter(x1,y1)
plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.subplot(221)
plt.scatter(x,y)

plt.subplot(222)
plt.scatter(x1,y1)

plt.show()
```
