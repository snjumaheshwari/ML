### Data For Classification
- Let us generate some examples for binary classification

```python
# Use multivariate normal distribution
# mean and covariance

u1=np.array([2,4])
cov_1=np.array([
    [1,0.5],
    [0.5,1]])
```

```python
x1= np.random.multivariate_normal(u1,cov_1,100)
plt.scatter(x1[:,0],x1[:,1])
plt.show()
```

```python
x1=np.random.multivariate_normal(u1,cov_1,300)
plt.scatter(x1[:,0],x1[:,1])
plt.show()
```

```python
print(x1.shape)
```

```python

u2=np.array([-1,-3])
cov_2=np.array([
    [3,-0.6],
    [-0.6,20
    ]])

x1=np.random.multivariate_normal(u1,cov_1,300)
plt.scatter(x1[:,0],x1[:,1], label="apples")

x2=np.random.multivariate_normal(u2,cov_1,400)
plt.scatter(x2[:,0], x2[:,1],label="oranges")

plt.title("Apples and oranges")
plt.xlabel("Size")
plt.ylabel("sweetness")
plt.legend()
plt.show()

```

```python
y=np.zeros(700)
y[:300]=1
print(y.shape)
```

```python
np.vstack?
```

```python
print(x1.shape,x2.shape)
X=np.vstack((x1,x2))
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
for i in range(700):
    if(y[i]==1):
        plt.scatter(X[i,0],X[i,1],color='green')
    else:
        plt.scatter(X[i,0],X[i,1],color='red')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

### Generate Data Using SkLearn

- Scikit-Learn is a powerful machine learning library
- pip install -U scikit-learn
- some useful functions to generate datasets available in sklearn.datasets
    - make_regression
    - make_classification
    - make_blobs
    - make_moons

```python
from sklearn.datasets import make_regression,make_classification
from matplotlib import pyplot as plt
```

```python
make_regression?
```

```python
X,Y =make_regression(n_samples=400,n_features=1,
                     n_informative=1, noise=4.8)
print(X.shape,Y.shape)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
make_classification?
```

```python
XC,YC= make_classification(n_samples=400,
                           n_features=2,n_informative=2,
                           n_redundant=0,
                           n_clusters_per_class=1,
                           random_state=6,
                           n_classes=3
                            )
```

```python
print(XC.shape,YC.shape)
```

```python
plt.scatter(XC[:,0], XC[:,1],c=YC)
plt.show()
```
